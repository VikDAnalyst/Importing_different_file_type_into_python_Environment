```python
#Import libraries
import numpy as np
import pandas as pd 
import json
from sas7bdat import SAS7BDAT
from sqlalchemy import create_engine, inspect
```

# Comma Separated Value File (.csv)

```python
# Importing CSV files using pandas
file="drake_data.csv"
drake_data_csv = pd.read_csv(file)
drake_data_csv.head()
```

```python
# Number of rows can also be specified 
drake_data_csv_nrows = pd.read_csv(file, nrows=100)
drake_data_csv_nrows.tail()
```

# Text File (.txt)

```python
# reading a TXT file in Python
with open("drake_lyrics.txt", "r", encoding="utf-8") as file:
    drake_data_txt=file.read()
    print(drake_data_txt)
```

# JavaScript Object Notation File (.json)

```python
#load JSON file into Python
file_path = "drake_data.json"
with open(file_path, "r", encoding="utf-8") as file:
    drake_lyrics_data = json.load(file)
type(drake_lyrics_data)
```

```python
# Since the data is a list data type, it can be converted into a DataFrame 
drake_lyrics_json = pd.DataFrame(drake_lyrics_data)
drake_lyrics_json.head()
```

# Statistcal Analysis System (SAS) File (.sas7bdat)

```python
# load sas file using pandas
sas_file = "naws_all.sas7bdat"
sas_df = pd.read_sas(sas_file, format="sas7bdat")
sas_df.head()
```
